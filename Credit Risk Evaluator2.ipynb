```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data

dependent_columns= ["loan_amnt","int_rate","installment", "home_ownership","annual_inc","verification_status", "pymnt_plan", "pct_tl_nvr_dlq", "percent_bc_gt_75", "pub_rec_bankruptcies","tax_liens","tot_hi_cred_lim",	"total_bal_ex_mort", "total_bc_limit", "total_il_high_credit_limit", "hardship_flag", "debt_settlement_flag"]

target= "loan_status"
```

```python
# Convert categorical da[ta to numeric and separate target feature for testing data
x_train = train_df.drop(['loan_status', 'Unnamed: 0', 'index'], axis=1)
y_train = train_df[target]

x_train_dummies=pd.get_dummies(x_train)
x_train_dummies.head()
```

```python
# add missing dummy variables to testing set
x_test = test_df.drop(['loan_status', 'Unnamed: 0', 'index'], axis=1)
y_test = test_df[target]

x_test_dummies=pd.get_dummies(x_test)
x_test_dummies.head()

```

```python
for i in x_train_dummies.columns:
    if i not in  x_test_dummies.columns:
        x_test_dummies[i]=0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.model_selection import train_test_split



from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

classifier

classifier.fit(x_train_dummies, y_train)
print(f"Training Data Score: {classifier.score(x_train_dummies, y_train)}")
print(f"Testing Data Score: {classifier.score(x_test_dummies, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score

clf = RandomForestClassifier(random_state=1, n_estimators=50).fit(x_train_dummies, y_train)
print(f'Training Score: {clf.score(x_train_dummies, y_train)}')
print(f'Testing Score: {clf.score(x_test_dummies, y_test)}')

```

```python
# Scale the data

scaler = StandardScaler().fit(x_train_dummies)
X_train_scaled = scaler.transform(x_train_dummies)
X_test_scaled = scaler.transform(x_test_dummies)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression()


classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf2 = RandomForestClassifier(random_state=1, n_estimators=50).fit(X_train_scaled, y_train)
print(f'Training Score: {clf2.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf2.score(X_test_scaled, y_test)}')
```
